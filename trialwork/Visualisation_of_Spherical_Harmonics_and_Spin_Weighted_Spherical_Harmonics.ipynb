```python
import waveformtools.transforms as wt
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
from waveformtools.transforms import Yslm_vec
```

```python
# Parameters
ell = 2     # Degree
m = 2       # Order
theta, phi = np.meshgrid(np.linspace(0, np.pi, 100), np.linspace(0, 2 * np.pi, 200))

# Calculate the spherical harmonics
Y = sph_harm(m, ell, phi, theta)

# Real and Imaginary components
real_part = np.real(Y)
imag_part = np.imag(Y)

# Create plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the real part
im1 = axes[0].imshow(real_part, extent=(0, 2 * np.pi, 0, np.pi), origin='lower', cmap='RdBu')
axes[0].set_title(f'Real part of SH({ell}, {m})')
fig.colorbar(im1, ax=axes[0], orientation='vertical')

# Plot the imaginary part
im2 = axes[1].imshow(imag_part, extent=(0, 2 * np.pi, 0, np.pi), origin='lower', cmap='RdBu')
axes[1].set_title(f'Imaginary part of SH({ell}, {m})')
fig.colorbar(im2, ax=axes[1], orientation='vertical')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm

# Parameters
ell = 2     # Degree
m = 2       # Order
s = -2      # Spin weight
theta, phi = np.meshgrid(np.linspace(0, np.pi, 100), np.linspace(0, 2 * np.pi, 200))

# To calculate the spin-weighted spherical harmonics
Y_s = sph_harm(m, ell, phi, theta)

# Spherical coordinates to Cartesian coordinates
x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

# To create the figure
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the SWSHs on the unit sphere
scatter = ax.scatter(x, y, z, c=np.real(Y_s), cmap='RdBu', s=10)

# Set plot labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Adding a colorbar
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Real part of Spherical Harmonics')

# Set aspect ratio
ax.set_box_aspect([1, 1, 1])

plt.show()
```
