# Function to produce visualisation of Ylm expanding in r_s

```python
def vis1Ylm_rs(ell, emm):
    import numpy as np
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D  
    from scipy.special import sph_harm
    import matplotlib
    
    # Spherical coordinates
    theta, phi = np.meshgrid(np.linspace(0, np.pi, 100), np.linspace(0, 2 * np.pi, 200))
    
    # Compute the spherical harmonics
    Y_lm = sph_harm(emm, ell, phi, theta).real
    
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    
    xr = Y_lm * x
    yr = Y_lm * y
    zr = Y_lm * z
    
    fig = plt.figure(figsize=(15, 11))
    ax = plt.axes(projection='3d')

    # Calculate color dimension
    U = Y_lm.real

    # Create colormap
    norm = plt.Normalize(U.min(), U.max())
    cmap = plt.cm.viridis

    # Plot the surface
    surface = ax.plot_surface(xr, yr, zr, rstride=1, cstride=1, linewidth=0, antialiased=False, facecolors=cmap(norm(U)))
    
    # Adding a colorbar
    # Create colorbar
    cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), ax=ax, pad=0.1)
    cbar.set_label('Real Part of Y_lm')
    
    # Set aspect ratio
    ax.set_box_aspect([1, 1, 1])
    
    ax.axis('off')
    
    plt.show()
```

# Example usage of the function visYlm()
