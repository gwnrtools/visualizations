# Visualisation function 1 of Yslm 

```python
def vis1Yslm (ell, emm):
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib
    from waveformtools.transforms import Yslm_vec
    from waveformtools.grids import UniformGrid
    from mpl_toolkits.mplot3d import Axes3D

    ginfo = UniformGrid()
    theta, phi = ginfo.meshgrid

    spin_weight = -2

    # Compute the spherical harmonics
    Y = Yslm_vec(spin_weight, ell, emm, theta, phi).real
    
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    
    xr = Y * x
    yr = Y * y
    zr = Y * z
    
    fig = plt.figure(figsize=(15, 11))
    ax = plt.axes(projection='3d')

    # Calculate color dimension
    U = Y.real

    # Create colormap
    norm = plt.Normalize(U.min(), U.max())
    cmap = plt.cm.viridis

    # Plot the surface
    surface = ax.plot_surface(xr, yr, zr, rstride=1, cstride=1, linewidth=0, antialiased=False, facecolors=cmap(norm(U)))
    
    # Adding a colorbar
    # Create colorbar
    cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), ax=ax, pad=0.1)
    cbar.set_label('Real Part of Y_slm')
    
    # Set aspect ratio
    ax.set_box_aspect([1, 1, 1])
    
    ax.axis('off')
    
    plt.show()
```
